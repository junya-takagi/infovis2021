```python
df = pd.read_csv("data.csv",encoding="utf-8")
df2= pd.read_csv("data2.csv",encoding="utf-8")
df.head()
```

```python
df = df[["都道府県","2018"]]
df.head()
```

```python
df2= df2[["都道府県","高齢化率"]]
df2.head()
```

```python
mdf = pd.merge(df,df2,on="都道府県")
```

```python
mdf.rename(columns={"都道府県":"prefecture","2018":"gdp","高齢化率":"aging_rate"},inplace=True)
```

```python
mdf.head()
```

```python
gdp = mdf["gdp"]/100/10000
mdf.index.name = "index"
gdp = gdp.round(2)
mdf["gdp"] = gdp
```

```python
mdf["region"] = 0
mdf["area"] = ""
n = len(mdf["prefecture"])
for i in range(n):
    if mdf["prefecture"][i] in ("北海道","青森県","岩手県","宮城県","秋田県","山形県","福島県"):
        mdf["area"][i] = "東北"
    elif mdf["prefecture"][i] in ("茨城県","栃木県","群馬県","埼玉県","千葉県","神奈川県"):
        mdf["area"][i] = "関東"
    elif mdf["prefecture"][i] in ("新潟県","富山県","福井県","長野県","岐阜県","静岡県","山梨県","愛知県"):
        mdf["area"][i] = "中部"
    elif mdf["prefecture"][i] in ("三重県","大阪府","兵庫県","和歌山県","滋賀県","京都府"):
        mdf["area"][i] = "近畿"
    elif mdf["prefecture"][i] in ("鳥取県","島根県","岡山県","広島県","山口県"):
        mdf["area"][i] = "中国"
    elif mdf["prefecture"][i] in ("高知県","徳島県","愛媛県","香川県"):
        mdf["area"][i] =  "四国"
    else:
        mdf["area"][i] = "九州"
mdf
```

```python
mdf.to_csv("plot_data.csv",encoding="utf-8")
```

```python
from sklearn.cluster import KMeans
import numpy as np
import matplotlib.pyplot as plt
```

```python
data_array = np.array([mdf["gdp"].tolist(),mdf["aging_rate"].tolist()])
data_array = data_array.T
```

```python
n = 4
pred = KMeans(n_clusters=n).fit_predict(data_array)
pred
```

```python
mdf["cluster_id"] = pred
```

```python
colors = ["red","blue","yellow","green"]
for i in range(0,n):
    data = mdf[mdf["cluster_id"]==i]
    xvalue = data["gdp"]
    yvalue = data["aging_rate"]
    plt.scatter(xvalue,yvalue,color=colors[i])
```
